```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
values_y = np.array([2,5,8,5])
plt.plot(values_y)
plt.show()
```

```python
values_x = np.array([1,2,3,4,5])
values_y = np.array([2,10,4,15,20])
plt.plot(values_x, values_y)
plt.show()
```

```python
#square green (sg) blue(sb)
plt.plot(values_x, values_y, "sb")
plt.show()
```

```python
plt.plot(values_x, values_y, "--", linewidth=1)
plt.show()
```

```python
x1 = np.random.randint(low=1, high=100, size=10)
y1 = np.random.randint(low=1, high=100, size=10)
plt.plot(x1, y1, 'co')

x2 = np.random.randint(low=1, high=100, size=10)
y2 = np.random.randint(low=1, high=100, size=10)
plt.plot(x2, y2, 'vg')

plt.show()
```

```python
y3 = np.random.randn(1000)
plt.hist(y3)
plt.show()
```

```python
#plt.hist(y3, orientation='horizontal', color='#CCCCCC' )
plt.hist(y3, orientation='horizontal', color='red' )
plt.show()
```

```python
#plt.hist(y3, range=(-1,1))
plt.hist(y3, range=(-2,2))
plt.show()
```

```python
x4 = np.random.randn(1000)
y4 = np.random.randn(1000)
plt.hist2d(x4, y4)
plt.show()
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
x5 = ['a','b','c','d','e']
y5 = np.random.randint(low=1,high=20, size=5)
ax.bar(x5, y5)
plt.show()
```

```python
#data
n = 4
y6_1 = (90, 55, 40, 65)
y6_2 = (85, 62, 54, 20)

#plot
fig, ax = plt.subplots()
index = np.arange(n)
bar_w = 0.35

#barras1
rects1 = plt.bar(index, y6_1, bar_w, color = 'r', label="valores 1")

#barras2
rects2 = plt.bar(index + bar_w, y6_2, bar_w, color = 'g', label="valores 2")

#titulos y leyendas
plt.xlabel("etiquetas")
plt.ylabel("valores")
plt.title("titulo del grafic")
plt.xticks(index + bar_w, ('A', 'B', 'C', 'D'))
plt.legend()

plt.tight_layout()
plt.show()

```

```python
v1 = np.array([0.2, 0.34, 0.1, 0.15, 0.05, 0.16 ])
plt.pie(v1)
plt.show()
```
